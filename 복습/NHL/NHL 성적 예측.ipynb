```python
# sklearn.metrics를 이용하여 모델 평가 수행
from sklearn.metrics import mean_absolute_error, r2_score, explained_variance_score, mean_squared_error

for index, row in combined_df1.iterrows():
    # 해당 연도의 예측값
    predicted_values = row[['G_82', 'A_82', 'PTS_82', '+/-_82', 'PIM_82', 'PPG_82', 'GWG_82', 'PP_82', 
                            'SOG_82', 'SPCT_82', 'TSA_82', 'TOI_82', 'FOW_82', 'FO%_82', 'BLK_82', 'HIT_82']]
    
    # 해당 연도의 실제값
    actual_values = df3.loc[index, ['G', 'A', 'PTS', '+/-', 'PIM', 'PPG', 'GWG', 'PP', 
                                     'SOG', 'SPCT', 'TSA', 'TOI', 'FOW', 'FO%', 'BLK', 'HIT']]
    
    # 모델 예측값과 실제값을 비교하여 정확도를 평가합니다.
    mae = mean_absolute_error(actual_values, predicted_values)
    r2 = r2_score(actual_values, predicted_values)
    evs = explained_variance_score(actual_values, predicted_values)
    mse = mean_squared_error(actual_values, predicted_values)
    
    print(f"Model {row['Model']}: MAE = {mae:.2f}, R2 = {r2:.2f},  MSE = {mse:.2f}, Explained Variance Score = {evs:.2f}")
```
